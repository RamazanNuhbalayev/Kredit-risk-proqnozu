## Məzmun
Kateqoraların və simvolların mürəkkib sistemi səbəbindən əsl məlumat bazasını başa düşmək demək olar imkansızdır. Buna görə də, onu oxunaqlı bir CSV fayına çevirmək üçün kiçik bir Python skript yazdım. Bir neçə sütun sadəcə göz ardı edilir, çünki mənim fikrimcə ya onlar vacib deyil ya da onların təsvirləri qəflətidir. The selected attributes are:

<br>**Age** (ədədi)
<br>**Sex** (mətn: kişi, qadın)
<br>**Job** (ədədi: 
* 0 – Unskilled and non-resident (İxtisassız və qeyri-rezident): Peşə və ya bacarıq tələb etməyən işlərdə çalışan və həmin ölkənin/sahənin daimi sakini olmayan şəxs.
* 1 – Unskilled and resident (İxtisassız və rezident): Peşə və ya bacarıq tələb etməyən işlərdə çalışan, lakin yaşadığı yer üzrə daimi sakindir.
* 2 – Skilled (İxtisaslı): Müəyyən peşə və ya texniki bacarıqlara malik olan işçi (məsələn, texnik, ustalar, sürücülər və s.).
* 3 – Highly skilled (Yüksək ixtisaslı): Ali təhsil və ya yüksək ixtisas tələb edən peşələrdə çalışan şəxslər (məsələn, həkimlər, mühəndislər, proqramçılar və s.))

<br>**Housing** (mətn: öz mülk, kirə və ya pulsuz)
<br>**Saving accounts** (mətn - az, orta, olduqca zəngin, zəngin)
<br>**Checking account** (ədədi, DM-də - Deutsch Mark)
<br>**Credit amount** (ədədi, DM-də)
<br>**Duration** (ədədi, ayda)
<br>**Purpose** (mətn: avtomobil, mebel/avadanlıq, radio/TV, məişət cihazları, təmir, təhsil, biznes, istirahət/digərləri)
<br>**Risk** (Hədəf dəyişən - Good Risk və ya Bad Risk)

```python
df_good = df_credit.loc[df_credit["Risk"] == 'good']['Age'].values.tolist()
df_bad = df_credit.loc[df_credit["Risk"] == 'bad']['Age'].values.tolist()
df_age = df_credit['Age'].values.tolist()

#First plot
trace0 = go.Histogram(
    x=df_good,
    histnorm='probability',
    name="Good Credit"
)
#Second plot
trace1 = go.Histogram(
    x=df_bad,
    histnorm='probability',
    name="Bad Credit"
)
#Third plot
trace2 = go.Histogram(
    x=df_age,
    histnorm='probability',
    name="Overall Age"
)

#Creating the grid
fig = tls.make_subplots(rows=2, cols=2, specs=[[{}, {}], [{'colspan': 2}, None]],
                          subplot_titles=('Good','Bad', 'Ümumi paylanma'))

#setting the figs
fig.append_trace(trace0, 1, 1)
fig.append_trace(trace1, 1, 2)
fig.append_trace(trace2, 2, 1)

fig['layout'].update(showlegend=True, title='Yaş sütunun statistik olaraq paylanması', bargap=0.05)
py.iplot(fig, filename='custom-sized-subplot-with-subplot-titles')
```

<h2>Creating an categorical variable to handle with the Age variable </h2>

```python
# Yaş aralıqları təyin olunur: 18-25, 25-35, 35-60, 60-120
interval = (18, 25, 35, 60, 120)

# Bu aralıqlara uyğun olaraq kateqoriyalar təyin olunur
cats = ['Tələbə', 'Gənc', 'Yetişkin', 'Yaşlı']
df_credit["Age_cat"] = pd.cut(df_credit.Age, interval, labels=cats)

df_good = df_credit[df_credit["Risk"] == 'good']
df_bad = df_credit[df_credit["Risk"] == 'bad']
```

```python
import plotly.graph_objs as go
import plotly.offline as py

# Job dəyərləri üçün kateqoriya adları
job_labels = {
    0: "İxtisassız və qeyri-rezident",
    1: "İxtisassız və rezident",
    2: "İxtisaslı",
    3: "Yüksək ixtisaslı"
}

# Good credit üçün saylar və kateqoriya adları
job_good = df_credit[df_credit["Risk"]== 'good']["Job"].value_counts().sort_index()
job_good.index = job_good.index.map(job_labels)

# Bad credit üçün saylar və kateqoriya adları
job_bad = df_credit[df_credit["Risk"]== 'bad']["Job"].value_counts().sort_index()
job_bad.index = job_bad.index.map(job_labels)

# First plot
trace0 = go.Bar(
    x = job_good.index,
    y = job_good.values,
    name = 'Good credit Distribution',
    marker_color='royalblue'
)

# Second plot
trace1 = go.Bar(
    x = job_bad.index,
    y = job_bad.values,
    name = 'Bad Credit Distribution',
    marker_color='tomato'
)

data = [trace0, trace1]

layout = go.Layout(
    title='İxtisaslaşma üzrə Kredit Riskinin Paylanması',
    barmode='group'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='grouped-bar')
```

## Bu dəyişənlər üzərində xüsusiyyət mühəndisliyi (feature engineering) aparaq və onların dəyərləri əsasında dummy (sıfır-bir) dəyişənlər yaradaq.

```python
def one_hot_encoder(df, nan_as_category = False):
    original_columns = list(df.columns)
    categorical_columns = [col for col in df.columns if df[col].dtype == 'object']
    df = pd.get_dummies(df, columns= categorical_columns, dummy_na= nan_as_category, drop_first=True)
    new_columns = [c for c in df.columns if c not in original_columns]
    return df, new_columns
```

## Məlumatların Dummy Dəyişənlərə Çevrilməsi

```python
df_credit['Saving accounts'] = df_credit['Saving accounts'].fillna('no_inf')
df_credit['Checking account'] = df_credit['Checking account'].fillna('no_inf')

#Purpose to Dummies Variable
df_credit = df_credit.merge(pd.get_dummies(df_credit.Purpose, drop_first=True, prefix='Purpose'), left_index=True, right_index=True)
#Sex feature in dummies
df_credit = df_credit.merge(pd.get_dummies(df_credit.Sex, drop_first=True, prefix='Sex'), left_index=True, right_index=True)
# Housing get dummies
df_credit = df_credit.merge(pd.get_dummies(df_credit.Housing, drop_first=True, prefix='Housing'), left_index=True, right_index=True)
# Housing get Saving Accounts
df_credit = df_credit.merge(pd.get_dummies(df_credit["Saving accounts"], drop_first=True, prefix='Savings'), left_index=True, right_index=True)
# Housing get Risk
df_credit = df_credit.merge(pd.get_dummies(df_credit.Risk, prefix='Risk'), left_index=True, right_index=True)
# Housing get Checking Account
df_credit = df_credit.merge(pd.get_dummies(df_credit["Checking account"], drop_first=True, prefix='Check'), left_index=True, right_index=True)
# Housing get Age categorical
df_credit = df_credit.merge(pd.get_dummies(df_credit["Age_cat"], drop_first=True, prefix='Age_cat'), left_index=True, right_index=True)
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score # to split the data
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, fbeta_score #To evaluate our model

from sklearn.model_selection import GridSearchCV

# Algorithmns models to be compared
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from xgboost import XGBClassifier
```

```python
df_credit['Credit amount'] = np.log(df_credit['Credit amount'])
```

```python
# X və y dəyişənlərinin yaradılması
X = df_credit.drop(columns='Risk_bad').values
y = df_credit["Risk_bad"].values

# Təlim və test dəstlərinə bölünməsi
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
# Təkrarlanabilirlik üçün toxum (seed)
seed = 7

# Modellərin siyahısı
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('RF', RandomForestClassifier()))
models.append(('SVM', SVC(gamma='auto')))
models.append(('XGB', XGBClassifier()))

# Hər bir modeli qiymətləndir
results = []
names = []
scoring = 'recall'  # Əsas göstərici: Recall (Geri çağırış)

for name, model in models:
    kfold = KFold(n_splits=10, shuffle=True, random_state=seed)
    cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)

# Nəticələrin qrafiklə müqayisəsi
fig = plt.figure(figsize=(11, 6))
fig.suptitle('Alqoritmlərin Müqayisəsi (Recall göstəricisinə görə)')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.ylabel('Recall')
plt.show()
```

Əldə olunan nəticələrə əsasən, əksər modellərdə geri çağırış (recall) göstəricisi aşağıdır. Yalnız Naive Bayes modeli 0.6-dan yuxarı nəticə göstərərək fərqlənir. CART və XGBoost modelləri də digər modellərlə müqayisədə daha stabil və qənaətbəxş recall dəyərlərinə malikdir. SVM isə ən zəif nəticəni vermişdir.
